```python
# import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read the csv file 
df = pd.read_csv('comp2_iqr.csv', index_col=0)
df.head()
```

```python
# get a list of the column names 
# will need this later in the notebook
df.columns
```

```python
# import the preprocessing package from sklearn
from sklearn import preprocessing

# define column names
cols = ['X1','X12_abs','X13_abs','X14_abs','X15_abs','X16_abs','X17_abs']

# scale all columns to have a range of 0-1
x = df.values
scaler = preprocessing.MinMaxScaler()
x_scaled = scaler.fit_transform(x)
df_scaled = pd.DataFrame(x_scaled,columns=cols)
df_scaled.head()
```

```python
# verify that the min and max values for each column are 0 and 1 respectively 
df_scaled.describe()
```

```python
# export new dataframe to csv
df_scaled.to_csv('comp2_minmax.csv')
```
