# **Exploratory Data Analysis**

```python
# Imported the needed packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
# Loaded in the data
df = pd.read_csv('../csv_files/comp2_raw.csv', header=0, index_col=0)
df.head(5)
```

```python
# Looking at the different columns that we have
df.columns
```

```python
# Seeing how many rows and entries that we have
df.shape
```

```python
# loooking for missing values - No missing values! 
print(df.isnull().sum(axis=0).tolist())
```

```python
# verifying no missing values
df.count()
```

```python
# Looking at the descriptive statisticss for each column
df.describe(include="all")
```

```python
# may want to bin
df['X2'].value_counts().plot(kind='bar', title="Gender Counts (1=Male, 2=Female)")
plt.show()
```

```python
# Using a bar chart to see what the general eduaction level is of the people in debt
# may want to bin - don't know what 5, 6, and 0 represent
df['X3'].value_counts().plot(kind='bar', title="Education Level (1=Graduate School, 2=University, 3=High School)")
plt.show()
```

```python
# Using a bar chart to see general relationship status of individuals in our data
# also may want to bin
df['X4'].value_counts().plot.bar(title="Marital Status (1=Married, 2=Single, 3=Other)")
plt.savefig('maritalstatus.png')
```

```python
# Examining the Age feature
# Highly skewed - may be a good candidate for binning later on
plt.hist(df['X5'])
plt.title('Age Distribution')
plt.show()
```

### Checking the Balance of the Target

```python
# Using a bar chart to display amount of individuals who defaulted and who did not 
# clearly not balanced - will have to address before modeling stage 
df['Y'].value_counts().plot.bar(title="Default? (0=No, 1=Yes)")
plt.savefig('y_imbalanced.png')
```

```python
# Plotting a histogram of columns that have nominal values. 
### Used the website at the bottom for help with the code
col_names = ['X1','X5', 'X6', 'X7','X8', 'X9', 'X10', 'X11', 'X12', 'X13', 'X14', 'X15', 'X16', 'X17', 'X18', 'X19', 'X20', 'X21', 'X22', 'X23'] 

fig, ax = plt.subplots(len(col_names), figsize=(10,100))

for i, col_val in enumerate(col_names):

    sns.distplot(df[col_val], hist=True, ax=ax[i])

plt.show()
```

```python
#Created a heatmap to see the correlation between the different variables
#Created a figure and are able to adjust the size of the graph
plt.figure(figsize = (20,20))

#Created a heatmap using Seaborn to display the correlation based on 'Spearman' method
##Added the annotation so that the numbers appear in each box
display = sns.heatmap(df.corr(method = 'spearman'), annot=True, fmt='.2f', cmap='BuPu', linewidth=.75, cbar_kws={'shrink': .75})

#Created a title
plt.title('Correlation Heatmap')
plt.savefig('corr_raw.png')
```
