```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('pipeline_2/csv_files/p2_comp2_final.csv', index_col=0)
df.head()
```

```python
df_shuffle=df.iloc[np.random.permutation(len(df))]
credit=df_shuffle.reset_index(drop=True)
credit.head()
```

```python
credit['Y']=credit['Y']
credit.head()
```

```python
credit_class = credit['Y'].values
```

```python
pd.isnull(credit).any()
```

```python
from sklearn.model_selection import train_test_split
training_indices, testing_indices = train_test_split(credit.index,stratify = credit_class,train_size=0.75, test_size=0.25, random_state = 2019)
training_indices.size, testing_indices.size
```

```python
from tpot import TPOTClassifier
from tpot import TPOTRegressor

tpot = TPOTClassifier(verbosity=2, max_time_mins=2, max_eval_time_mins=0.04, population_size=15)
tpot.fit(credit.drop('Y',axis=1).loc[training_indices].values, credit.loc[training_indices,'Y'].values)
```

```python
tpot.export('tpot_credit_pipeline3.py')
```
